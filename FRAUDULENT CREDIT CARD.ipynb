```python
#seperating the fraudulent card from the normal ones
filt_fraudulent = credit_card["Class"] ==1
filt_normal = credit_card["Class"] == 0
```

```python
#view the fraudulent cards
fraudulent =credit_card.loc[filt_fraudulent]
fraudulent
```

```python
#view the normal ones
normal =credit_card.loc[filt_normal]
normal
```

```python
# sort the data according to the amount(smallest to largest)
credit_card.sort_values(by="Amount", inplace=True)
credit_card
```

# Data Visualization

```python
data_y = [fraudulent["Class"].count(), normal["Class"].count()]
data_x = [1,0]

colors= ["red", "green"]
plt.bar(data_x, data_y, log=True, color=colors)


plt.xlabel("Class")
plt.ylabel("Number of transactions")
plt.xticks(ticks=data_x, labels=["fraudulent", "normal"])
plt.show()
```

```python
bins= 50
plt.hist(normal["Amount"], bins=bins, color="green")
plt.title("Amount per transactions for Normal cards")
plt.yscale("log")
plt.xlabel("Amount")
plt.ylabel("number of transactions")
plt.xlim((0,20000))
plt.show()
```

```python
bins= 50
plt.hist(fraudulent["Amount"], bins=bins)
plt.title("Amount per transactions for fraudulent cards")
plt.yscale("log")
plt.xlim((0,20000))
plt.xlabel("Amount")
plt.ylabel("number of transactions")
plt.show()
```

```python
# top 10 fraudulent transactions based on amount
fraudulent = fraudulent.sort_values(by="Amount")
y = fraudulent["Amount"].tail(10)
x= [i for i in range(10,0,-1)]
plt.xlabel("Positions")
plt.ylabel("Amount")
plt.title("Top 10 fraudulent transactions based on amount")
plt.bar(x,y)
plt.show()
```
